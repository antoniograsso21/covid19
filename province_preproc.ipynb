```python
import time
import requests
import datetime
import pandas as pd

from util.functions import Functions

# Disable warning
pd.options.mode.chained_assignment = None 

level = 'province'
path_data = './data'
path_csv = path_data + '/csv'
file_name_out_p = path_csv + '/covid19-' + level + '.csv'
```

```python
# Define today, yesterday and day before yesterday dates
date_tday = datetime.date.today()
date_yday = Functions.get_previous_date(date_tday)
date_bf_yday = Functions.get_previous_date(date_yday)
date_tday, date_yday, date_bf_yday
```

```python
# Today url for regions file
url_tday = Functions.get_url_date(level, date_tday)
# Yesterday url for regions file
url_yday = Functions.get_url_date(level, date_yday)
# Day before yesterday url for regions file
url_bf_yday = Functions.get_url_date(level, date_bf_yday)

url_tday, url_yday, url_bf_yday
```

```python
# Get csv files
try:
    # today and yesterday if today present
    df_raw_p_1 = pd.read_csv(url_tday)
    print('Dati odierni')
    df_raw_p_0 = pd.read_csv(url_yday)
    date_last = date_tday
except Exception as e:
    # yesterday and day before yesterday otherwise
    df_raw_p_1 = pd.read_csv(url_yday)
    df_raw_p_0 = pd.read_csv(url_bf_yday)
    date_last = date_yday
    print('Dati di ieri')
```

```python
# State S
df_raw_p_1.sort_values(by='codice_provincia').reset_index(drop=True)
```

```python
# State S - 1
df_raw_p_0.sort_values(by='codice_regione').reset_index(drop=True)
```

```python
# Print raw columns
# print(df_raw_p_1.columns, df_raw_p_0.columns)
# Drop useless columns
cols_to_drop = ['stato', 'lat', 'long', 'note']
df_pf_1 = df_raw_p_1.drop(cols_to_drop, axis=1).sort_values(by='codice_provincia').reset_index(drop=True)
df_pf_0 = df_raw_p_0.drop(cols_to_drop, axis=1).sort_values(by='codice_provincia').reset_index(drop=True)
# df_pf_1.columns, df_pf_0.columns
```

```python
df_p = df_pf_1.copy()
df_p['variazione_totale_casi'] = df_pf_1['totale_casi'] - df_pf_0['totale_casi']
df_p
```

```python
df_p.to_csv(file_name_out_p, index=False)
```
