```python
import sys
import datetime
import numpy as np
import pandas as pd

path_root = '../..'
sys.path.append(f'{path_root}/')
from src.util.functions import Functions as F

# Disable warning
pd.options.mode.chained_assignment = None 

level = 'andamento-nazionale'

path_data = f'{path_root}/data'
path_csv = f'{path_data}/csv'
csv_out_name = f'{path_csv}/covid19-{level}.csv'
```

```python
url_cum = F.get_url_cum(level)
url_cum
```

```python
df_n_raw = pd.read_csv(url_cum)
df_n_raw
```

```python
# Drop unused columns
df_nd = df_n_raw.drop(['stato', 'note', 'note_test', 'note_casi'], axis=1)
df_nd
```

```python
list(df_nd.columns)
```

```python
# Add columns
df_nd['variazione_terapia_intensiva'] = df_nd['terapia_intensiva'].diff()
df_nd['variazione_ricoverati_con_sintomi'] = df_nd['ricoverati_con_sintomi'].diff()
df_nd['variazione_totale_ospedalizzati'] = df_nd['totale_ospedalizzati'].diff()
df_nd['variazione_isolamento_domiciliare'] = df_nd['isolamento_domiciliare'].diff()
df_nd['nuovi_dimessi_guariti'] = df_nd['dimessi_guariti'].diff()
df_nd['nuovi_deceduti'] = df_nd['deceduti'].diff()
df_nd['nuovi_tamponi'] = df_nd['tamponi'].diff()
df_nd['nuovi_casi_testati'] = df_nd['casi_testati'].diff()
df_nd
```

```python
df_n = df_nd.iloc[1:]
df_n
```

```python
# Add columns
df_n['ti_su_to'] = ((df_n['terapia_intensiva'] / df_n['totale_ospedalizzati']) * 100).round(2)
df_n['to_su_tp'] = ((df_n['totale_ospedalizzati'] / df_n['totale_positivi']) * 100).round(2)
df_n['np_su_nt'] = ((df_n['nuovi_positivi'] / df_n['nuovi_tamponi']) * 100).round(2)
df_n['np_su_nct'] = ((df_n['nuovi_positivi'] / df_n['nuovi_casi_testati']) * 100).round(2)
df_n['var_ti_perc'] = ((df_n['variazione_terapia_intensiva'] / (df_n['terapia_intensiva'] - df_n['variazione_terapia_intensiva'])) * 100).round(2)
df_n['var_pos_perc'] = ((df_n['variazione_totale_positivi'] / (df_n['totale_positivi'] - df_n['variazione_totale_positivi'])) * 100).round(2)
df_n['np_log'] = np.log10(df_n['nuovi_positivi'])
df_n
```

```python
df_n.to_csv(csv_out_name, index=False)
```
